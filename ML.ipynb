```python
import time

import numpy as np
np.set_printoptions(threshold=np.nan,suppress=True)
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
```

```python
def results(labels_test,predict_labels):
    tn, fp, fn, tp = confusion_matrix(labels_test,predict_labels).ravel()
    print('tn:',tn,'fp:',fp,'fn:',fn,'tp:',tp)
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    print("accuracy :",accuracy)
```

```python
def gen_clfs():
    clfs = {
    'CART': DecisionTreeClassifier(),
    'KNN2': KNeighborsClassifier(n_neighbors=2),
    'KNN5': KNeighborsClassifier(n_neighbors=5),
    'NBS': GaussianNB(),
    'ID3': DecisionTreeClassifier(criterion="entropy"),
    'RF10': RandomForestClassifier(n_estimators=10),
    'RF100': RandomForestClassifier(n_estimators=100),
    'B': BaggingClassifier(n_estimators=50),
    'AB': AdaBoostClassifier(n_estimators=300),
    'MLP': MLPClassifier()
    }
    return clfs

"""

    'MLP': MLPClassifier(hidden_layer_sizes=(100, ),
                         activation='relu',
                         solver='adam',
                         alpha=0.0001,
                         batch_size='auto',
                         learning_rate='constant',
                         learning_rate_init=0.001,
                         power_t=0.5,
                         max_iter=200,
                         shuffle=True,
                         random_state=None,
                         tol=0.0001,
                         verbose=False,
                         warm_start=False,
                         momentum=0.9,
                         nesterovs_momentum=True,
                         early_stopping=False,
                         validation_fraction=0.1,
                         beta_1=0.9,
                         beta_2=0.999,
                         epsilon=1e-08,
                         n_iter_no_change=10),   
"""

def gen_best_clfs():
    clfs = {
    'NBS': GaussianNB(),
    'AB': AdaBoostClassifier(base_estimator=None,
                             n_estimators=500,
                             learning_rate=1,
                             algorithm='SAMME.R',
                             random_state=None)
    }
    return clfs

def run_classifiers(dict_clf, X, labels):
    kf = KFold(n_splits=10, shuffle=True, random_state=0)
    for i in dict_clf:
        clf = dict_clf[i]
        t0 = time.time()
        cv_acc = cross_val_score(clf, X, labels, cv=kf)
        print("Execution time for {0} is: {1:.3f}s".format(i,time.time()-t0))
        print("Accuracy for {0} is: {1:.3f} +/-{2:.3f}".format(i, np.mean(cv_acc), np.std(cv_acc)))
        print()
```

```python
run_classifiers(gen_best_clfs(), X, labels)
```
